### Basic DHRobot

```python
puma560_dh_parameters =[[0,     0,      0], # unused dummy link 0
                        [0.6718,0,      pi/2],
                        [0,     0.4318, 0],
                        [0.15,  0.0203, -pi/2],
                        [0.4318,0,      pi/2],
                        [0,     0,      -pi/2],
                        [0,     0,      0]]

links = []
#links.append(rtb.RevoluteDH(d=puma560_dh_parameters[0][0], a=puma560_dh_parameters[0][1], alpha=puma560_dh_parameters[0][2]) )
links.append( rtb.RevoluteDH(d=puma560_dh_parameters[1][0], a=puma560_dh_parameters[1][1], alpha=puma560_dh_parameters[1][2]) )
links.append( rtb.RevoluteDH(d=puma560_dh_parameters[2][0], a=puma560_dh_parameters[2][1], alpha=puma560_dh_parameters[2][2]) )
links.append( rtb.RevoluteDH(d=puma560_dh_parameters[3][0], a=puma560_dh_parameters[3][1], alpha=puma560_dh_parameters[3][2]) )
links.append( rtb.RevoluteDH(d=puma560_dh_parameters[4][0], a=puma560_dh_parameters[4][1], alpha=puma560_dh_parameters[4][2]) )
links.append( rtb.RevoluteDH(d=puma560_dh_parameters[5][0], a=puma560_dh_parameters[5][1], alpha=puma560_dh_parameters[5][2]) )
links.append( rtb.RevoluteDH(d=puma560_dh_parameters[6][0], a=puma560_dh_parameters[6][1], alpha=puma560_dh_parameters[6][2]) )

puma560 = rtb.DHRobot(links, name='Puma 560')
```

### Adding other parameters

```python
deg_to_rad = pi/180

# Define the joint angle limits for the Puma 560 robot
puma560.qlim = array([
    [-160,  -45, -225, -110, -100, -266],
    [ 160,  225,   45,  170,  100,  266]
    ]) * deg_to_rad

q_lim = puma560.qlim
print(q_lim)
```

```python
def get_random_q():
    q = array([
        random.uniform(q_lim[0][0], q_lim[1][0]),
        random.uniform(q_lim[0][1], q_lim[1][1]),
        random.uniform(q_lim[0][2], q_lim[1][2]),
        random.uniform(q_lim[0][3], q_lim[1][3]),
        random.uniform(q_lim[0][4], q_lim[1][4]),
        random.uniform(q_lim[0][5], q_lim[1][5])
        ])
    return q
```

```python
puma560.qr = array([0, pi/2, -pi/2, 0, 0, 0])
puma560.qz = zeros(6)
puma560.qn = array([0, pi/4, pi, 0, pi/4, 0])
puma560.qs = array([0, 0, -pi/2, 0, 0, 0])

puma560.qh = array([0, 0, pi/2, 0, 0, 0])

puma560.ru = array([0, 45, 180, 0, 45, 0]) * deg_to_rad
puma560.rd = array([0, -47.8, 5.39, -180, 47.6, 180]) * deg_to_rad
puma560.lu = array([152, -225, 5.39, 145, 55.8, 21.4]) * deg_to_rad
puma560.ld = array([152, -132, 180, 38.6, 49.3, 152]) * deg_to_rad

puma560.rnd = get_random_q()

puma560.addconfiguration("qr", puma560.qr) # ready pose, arm up
puma560.addconfiguration("qz", puma560.qz) # zero angles, arm down
puma560.addconfiguration("qn", puma560.qn) # nominal table top picking pose
puma560.addconfiguration("qs", puma560.qs) # straight and horizontal

puma560.addconfiguration("qh", puma560.qh) # horizontal, but not straight

puma560.addconfiguration("ru", puma560.ru) # right-handed, elbow up configuration
puma560.addconfiguration("rd", puma560.rd) # right-handed, elbow down configuration
puma560.addconfiguration("lu", puma560.lu) # left-handed, elbow up configuration
puma560.addconfiguration("ld", puma560.ld) # left-handed, elbow down configuration

puma560.addconfiguration("rnd", puma560.rnd) # random configuration
```

### Constructing the Puma560 robot using ETS
(Optional)

- [A Tutorial on Manipulator Differential Kinematics: Notebook 1 Manipulator Kinematics](https://github.com/jhavl/dkt/blob/main/Part%201/1%20Manipulator%20Kinematics.ipynb)
- [Introduction to Elementary Transform Sequences (ETS)](https://github.com/petercorke/robotics-toolbox-python/blob/master/notebooks/ets.ipynb)

```python
E1 = rtb.ET.Rz()
E2 = rtb.ET.tz(0.6718)
E3 = rtb.ET.Rx(pi/2)
E4 = rtb.ET.Rz()
E5 = rtb.ET.tx(0.4318)
E6 = rtb.ET.Rz()
E7 = rtb.ET.tz(0.15)
E8 = rtb.ET.tx(0.0203)
E9 = rtb.ET.Rx(-pi/2)
E10 = rtb.ET.Rz()
E11 = rtb.ET.tz(0.4318)
E12 = rtb.ET.Rx(pi/2)
E13 = rtb.ET.Rz()
E14 = rtb.ET.Rx(-pi/2)
E15 = rtb.ET.Rz()

ets = E1 * E2 * E3 * E4 * E5 * E6 * E7 * E8 * E9 * E10 * E11 * E12 * E13 * E14 * E15

puma560_ets = rtb.Robot(ets, name='Puma 560')

puma560_ets
```

- [What are Singularities in a Six-Axis Robot Arm?](https://mecademic.com/insights/academic-tutorials/what-are-singularities-6-axis-robot-arm/)
    - "The typical six-axis industrial robot arm has three types of robot singularity: wrist singularities, elbow singularities, and shoulder/overhead singularities. A wrist singularity occurs when the axes of joints 4 and 6 are coincident. An elbow singularity occurs when the wrist center lies in the same plane as the axes of joints 2 and 3. Finally, a shoulder/overhead singularity occurs when the wrist center lies in the plane passing through the axis of joint 1 and parallel to or passing through the axis of joint 2."

### 3 cases of wrist singularity, where q6 depends on q4

```python
singularity_list = ["qz", "qs", "qh"]

qz_singu = array([0, 0, 0, pi/2, 0, 0]) # case 1: zero configuration qz
qs_singu = array([0, 0, -pi/2, pi/2, 0, 0]) # case 2: straight and horizontal configuration qs
qh_singu = array([0, 0, pi/2, pi/2, 0, 0]) # case 3: horizontal, but not straight configuration qh

for config in singularity_list:
    m = puma560.manipulability(eval(f"puma560.{config}"), axes='all')
    J = puma560.jacob0(eval(f"puma560.{config}")).round(2)

    print(f"Manipulability for configuration {config}:", m)
    print(f"Jacobian for configuration {config}:")
    print(J)
```

```python
for config in singularity_list:
    qt = rtb.tools.trajectory.jtraj(eval(f"puma560.{config}"), eval(f"{config}_singu"), 50)
    puma560.plot(qt.q, backend="pyplot", movie=f'Media/Puma560 {config} Singularity.gif')
```

## Finding Workspace

- [A Tutorial on Manipulator Differential Kinematics: Notebook 5 Manipulator Performance Measures](https://github.com/jhavl/dkt/blob/main/Part%201/5%20Manipulator%20Performance%20Measures.ipynb)

### Generating a lot of end-effector positions from random joint angles

```python
n_points = 100000
n_points_sample = 10000

axes = [True, True, True, True, True, True] # [tx ty tz Rx Ry Rz]

list_q_point = array([get_random_q() for i in range(n_points)]) # List of joint angles

list_of_points = array([puma560.fkine(list_q_point[i]).t for i in range(n_points)]) # List of end effector positions

print(list_of_points)
```

### Sampling points

```python
def sample_points(list_of_points, n_points_sample):

    if n_points_sample > len(list_of_points):
        n_points_sample = len(list_of_points)
    elif n_points_sample < 0:
        n_points_sample = 0

    sample = array(list_of_points[:n_points_sample])
    return sample
```

```python
list_q_point_sample = sample_points(list_q_point, n_points_sample)
list_of_points_sample = sample_points(list_of_points, n_points_sample)

print(len(list_of_points_sample))
print(list_of_points_sample)
```

### Calculating Yoshikawa manipulability index

```python
# puma560.manipulability()

def manipulability(robot, q, axes):
    """
    Calculates the manipulability of the robot at joint configuration q

    :robot: A Robot object to find the manipulability of
    :q: The joint coordinates of the robot (ndarray)
    :axes: A boolean list which correspond with the Cartesian axes to
        find the manipulability of (6 boolean values in a list)
    """
    # calculate the Jacobian
    J = robot.jacobe(q)

    # only keep the selected axes of J
    J = J[axes, :]

    # calculate the manipulability
    m = sqrt(linalg.det(J @ J.T))

    return m
```

```python
def manipulability_list(robot, list_q, axes, n_points):
    m_list = array([manipulability(robot, list_q[i], axes) for i in range(n_points)])
    return m_list
```

```python
m_list = manipulability_list(puma560, list_q_point, axes, n_points)
m_list_sample = manipulability_list(puma560, list_q_point_sample, axes, n_points_sample)

print(m_list)
print(m_list_sample)
```

#### Filtering points where singularity chance is high: where m is close to 0

```python
m_threshold = 0.001

m_list_singu = [[i, m] for i, m in enumerate(m_list) if m < m_threshold]

points_singu = array([list_of_points[i] for (i, m) in m_list_singu])

# print(m_list_singu)
print(points_singu)
```

### Plot the points on a 3D plot, with the color representing the manipulability index

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = list_of_points[:,0]
y = list_of_points[:,1]
z = list_of_points[:,2]
c = m_list
cmap = plt.cm.RdPu
s = 1

scatter = ax.scatter(x, y, z, c=c, cmap=cmap, s=s)

# set zorders here
scatter.set_zorder(1)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_box_aspect([1,1,1])

cbar = fig.colorbar(scatter, ax=ax)
cbar.set_label('Manipulability', rotation=270)

plt.title('Workspace of the Puma560 Robot')

plt.savefig('Media/Puma560 Workspace.png')

plt.show()
```
